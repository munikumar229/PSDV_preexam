1. ### List Operations and Basic Statistics:
- Create a list of the first 20 natural numbers.
- Compute the sum of all even numbers in the list.
- Compute the product of all odd numbers in the list.
- Calculate the mean, median, and standard deviation of the list.


```python
import numpy as np

natural=[]
for i in range(1,21):
    natural.append(i)
print("Natural numbers = ",natural)

even_sum=0
odd_product=1
for i in natural:
  if i%2==0:
    even_sum+=i
  else:
    odd_product*=i
print("Sum of even natural numbers = ",even_sum)
print("Product of odd natural numbers = ",odd_product)

mean=np.mean(natural)
median=np.median(natural)
std=np.std(natural)
print("Mean = ",mean)
print("Median = ",median)
print("Standard Deviation = ",std)
```

2. ### Mutable vs. Immutable Data Types:
- Explain the difference between mutable and immutable data types in Python. Provide examples of each.

1. Mutable Data Types:
- These data types are those whose values can be modified after the object is created.
- Examples : list, dict, set
2. Immutable Data Types:
- These data types are those whose contents cannot be changed after the object is created.
- If we try to modify an immutable object, a new object is created instead, and the original object remains unchanged.
- Examples : int, float, str, tuple


3. ### Fibonacci Sequence:
- Write a Python function that takes a number n and returns a list of the first n Fibonacci numbers.
- Plot the Fibonacci sequence using a line graph.

```python
n=int(input("Enter the integer n : "))

fibonacci=[0,1]
for i in range(2,n):
  fibonacci.append(fibonacci[i-1]+fibonacci[i-2])
print("First n Fibonacci numbers : ",fibonacci)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(range(n), fibonacci, marker='o', linestyle='-', color='b', label='Fibonacci Sequence')
plt.title("Fibonacci Sequence")
plt.xlabel("Index")
plt.ylabel("Value")
plt.grid(True)
plt.legend()
plt.show()
```

9. ### Merging Dictionaries:
- Write Python code to merge two dictionaries. If a key is present in both, sum their values.

- Example:
-- dict1 = {'a': 10, 'b': 20}
-- dict2 = {'b': 5, 'c': 15}
-- Output: {'a': 10, 'b': 25, 'c': 15}



```python
dict1 = {'a': 10, 'b': 20}
dict2 = {'b': 5, 'c': 15}

merged_dict = {}

for key, value in dict1.items():
    merged_dict[key] = value
    if key in dict2:
        merged_dict[key] += dict2[key]

for key, value in dict2.items():
    if key not in merged_dict:
        merged_dict[key] = value
print(merged_dict)
```

10. ### Character Frequency:
- Write a function that accepts a string and:
- Converts all characters to lowercase.
- Counts the frequency of each character.
- Returns a dictionary with characters as keys and their counts as values.
- Visualize the character frequencies using a bar chart.

```python
def character_frequency(string):
  string=string.lower()
  string=string.replace(" ","")
  dict={}
  for i in string:
    if i in dict:
      dict[i]+=1
    else:
      dict[i]=1
  return dict
```

11. ### Remove Duplicates:
- Write a program to remove duplicate elements from a list while maintaining the order of elements.
- Calculate the mean and median of the unique elements.

```python
def duplicate(list):
  seen=set()
  unique=[]
  for i in list:
    if i not in seen:
      seen.add(i)
      unique.append(i)
  return unique

list=[1,2,4,5,2,4,3,3]
unique_list=duplicate(list)
print("List with duplicates : ",list)
print("List without duplicates : ",unique_list)
print("Mean : ",np.mean(unique_list))
print("Median : ",np.median(unique_list))

```

13. ### Sorting Dictionaries:
Write a function to sort a list of dictionaries by a key.

Example:

- data = [{'name': 'Alice', 'age': 25}, {'name': 'Bob', 'age': 22}]
- Sort by age → [{'name': 'Bob', 'age': 22}, {'name': 'Alice', 'age': 25}]

```python
def sort_dict_list(data, key):
    sorted_data = sorted(data, key=lambda x: x[key])
    return sorted_data

data = [{'name': 'Alice', 'age': 25}, {'name': 'Bob', 'age': 22}]
sorted_data = sort_dict_list(data, 'age')

print("Sorted data by age:", sorted_data)
```

14. ### Word Filtering and Sorting:
Write a program to:
- Split a string into a list of words.
- Remove all words that are shorter than 4 characters.
- Sort the remaining words alphabetically.
- Visualize the word lengths using a bar chart.

```python
def filter_and_sort(input_string):

    words = input_string.split()

    filtered_words = [word for word in words if len(word) >= 4]

    sorted_words = sorted(filtered_words)

    word_lengths = [len(word) for word in sorted_words]

    plt.bar(sorted_words, word_lengths)
    plt.xlabel('Words')
    plt.ylabel('Word Lengths')
    plt.title('Word Lengths After Filtering and Sorting')
    plt.xticks(rotation=45)
    plt.show()

    return sorted_words, word_lengths


input_string = "A man a plan a canal Panama"
sorted_words, word_lengths = filter_and_sort(input_string)

print("Sorted words after filtering:", sorted_words)
print("Word lengths:", word_lengths)
```

15. ### Anagram Finder:
- Write a Python function to find all anagrams of a word in a given list.
-- Input: word = "listen", words = ["enlist", "google", "inlets", "banana"]
-- Output: ["enlist", "inlets"]

```python
def find_anagrams(word, words):
    sorted_word = sorted(word)

    anagrams = [w for w in words if sorted(w) == sorted_word]

    return anagrams


word = "listen"
words = ["enlist", "google", "inlets", "banana"]

anagrams = find_anagrams(word, words)
print("Anagrams of '{}' in the list:".format(word), anagrams)
```

16. ### Set Operations:
- Given two lists, write Python code to find their intersection, union, and symmetric difference.
- Visualize the results using Venn diagrams.

```python
from matplotlib_venn import venn2

def set_operations_and_venn(list1, list2):
    set1 = set(list1)
    set2 = set(list2)


    intersection = set1 & set2
    union = set1 | set2
    symmetric_difference = set1 ^ set2

    print("Intersection:", intersection)
    print("Union:", union)
    print("Symmetric Difference:", symmetric_difference)

    venn2(subsets=(len(set1 - set2), len(set2 - set1), len(intersection)),
          set_labels=('List 1', 'List 2'))
    plt.title('Venn Diagram of Set Operations')
    plt.show()

    return intersection, union, symmetric_difference

list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]

intersection, union, symmetric_difference = set_operations_and_venn(list1, list2)

```

18. ### Word Count and Frequency:
Write a program to:
- Read a file and count the number of words in it.
- Find the most frequent word.
- Visualize the word frequencies using a bar chart.

```python
from collections import Counter
import re

def word_count_and_frequency(file_path):

    with open(file_path, 'r') as file:
        text = file.read()


    words = re.findall(r'\b\w+\b', text.lower())

    word_count = len(words)

    word_frequency = Counter(words)
    most_frequent_word, most_frequent_count = word_frequency.most_common(1)[0]

    plt.figure(figsize=(10, 6))
    plt.bar(word_frequency.keys(), word_frequency.values())
    plt.title('Word Frequency Distribution')
    plt.xlabel('Words')
    plt.ylabel('Frequency')
    plt.xticks(rotation=90)
    plt.show()

    return word_count, most_frequent_word, most_frequent_count

file_path = '/content/short'
word_count, most_frequent_word, most_frequent_count = word_count_and_frequency(file_path)

print(f"Total number of words: {word_count}")
print(f"Most frequent word: '{most_frequent_word}' with {most_frequent_count} occurrences")
```

19. ### Random Number Generation:

- Generate 100 random integers between 1 and 50.
- Save them in a text file, one per line.
- Read the file and compute the mean of the numbers.
- Visualize the distribution of the numbers using a histogram.

```python
import random

random_numbers = [random.randint(1, 50) for _ in range(100)]

with open("random_numbers.txt", "w") as file:
    for num in random_numbers:
        file.write(f"{num}\n")


with open("random_numbers.txt", "r") as file:
    numbers_from_file = [int(line.strip()) for line in file.readlines()]

mean = np.mean(numbers_from_file)

print(f"Mean of the numbers: {mean}")

plt.hist(numbers_from_file, bins=range(1, 52), edgecolor='black', alpha=0.7)
plt.title('Distribution of Random Numbers')
plt.xlabel('Number')
plt.ylabel('Frequency')
plt.xticks(range(1, 51))
plt.show()
```

22. ### Normalization:

```python
list=[10,20,30,40,50,60,70]
min_value=min(list)
max_value=max(list)
normalized_list=[]
for i in list:
  normalized_list.append((i-min_value)/(max_value-min_value))
print("Normalized list : ",normalized_list)
```

25. ### Leap Year Checker:
- Write a Python program that takes a year as input and checks if it is a leap year.

```python
def is_leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False

year = int(input("Enter a year: "))

if is_leap_year(year):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```
